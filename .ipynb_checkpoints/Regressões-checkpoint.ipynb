```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot
from sklearn import datasets
from sklearn.metrics import r2_score

# Carregar os dados
df = pd.read_csv('spotify_songs.csv', header=(0))  # substitua pelo caminho correto

# Remover valores nulos
df = df.dropna()
```

```python
data = df.to_numpy()
y = df['track_popularity'].to_numpy()
x = df['danceability'].to_numpy()
X = np.array(x).reshape(-1, 1)
Y = np.array(y)
modelo = LinearRegression().fit(X, Y)

# Mostrar os coeficientes
print("Interceptação:", modelo.intercept_)
print("Coeficiente angular:", modelo.coef_[0])
y_pred=modelo.predict(X)
R2 = r2_score(y, y_pred)
print('R2:', R2)
correlacao = df['danceability'].corr(df['track_popularity'])
print(f"Correlação entre dançabilidade e popularidade: {correlacao:.3f}")

# Plotar os dados e a reta ajustada
plt.figure(figsize=(9,8), facecolor='none')
ax = plt.gca() 
ax.set_facecolor("none")
plt.scatter(X, Y, label='Dados', color='#121212')
plt.plot(X, modelo.predict(X), color='#1DB954', label='Reta ajustada', linewidth=3)
plt.xlabel('Dançabilidade')
plt.ylabel('Popularidade')
plt.title('Regressão Linear: Popularidade vs Dançabilidade')
plt.savefig('grafico.png', dpi=300, bbox_inches='tight')
plt.show()



```

```python
import seaborn as sns
import statsmodels.formula.api as sm
df_edm = df[df['playlist_genre'] == 'edm']
model = sm.ols(formula='track_popularity ~ duration_ms+energy+danceability+loudness+speechiness+valence', data=df_edm)

print("Antes da analise do p-valor")
result = model.fit()
print(result.summary())
```
